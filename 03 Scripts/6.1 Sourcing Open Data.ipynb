### Consistency Checks: no duplicates, mixed-type data, missing
values, or outliers

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')
```

```python
# Check for mixed-type data
mixed_type_columns = df.select_dtypes(include='object').columns[df.select_dtypes(include='object').apply(lambda x: x.nunique() < len(x))].tolist()
print(f'Mixed-type columns: {mixed_type_columns}')
```

```python
#To understand the nature of the mixed types, I'll examine the unique data types in these columns:

for col in mixed_type_columns:
    print(f"Column: {col}")
    print(df[col].apply(type).value_counts())
    print("\n")
```

```python
#Summary of Mixed-Type Data

for col in mixed_type_columns:
    unique_values = df[col].unique()
    print(f"Unique values in '{col}': {unique_values[:10]}")  # Display first 10 unique values
```

```python
# Check for missing values

missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values[missing_values > 0])
```
